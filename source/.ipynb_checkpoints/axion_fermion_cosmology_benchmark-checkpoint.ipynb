# Axion Production Rate
Our calculation of the axion production rate $\Gamma_a^{2\to2}$ through axion-fermion interactions.

Detailed notes see https://www.overleaf.com/read/mprpvjrnjzrg


## Compton Process $\gamma + \psi \to a + \psi$
We calculate the production rate derived in the note:
\begin{equation}
    \begin{split}
        \frac{n_a^{eq}}{\alpha |\tilde{\epsilon}|^2 T^4}\Gamma_{\gamma \psi \to \phi \psi} =& \frac{1}{8\pi^4 T^4}\int_{m_\psi}^\infty dE\int_0^{\sqrt{E^2 - m_\psi^2}} dp\int_\frac{E^2-m_\psi^2-p^2}{2(E+p)}^\frac{E^2-m_\psi^2-p^2}{2(E-p)} dp_1 \int_\frac{E^2-m_\psi^2-p^2}{2(E+p)}^\frac{E^2-m_\psi^2-p^2}{2(E-p)} dp_3 \frac{1}{(e^\frac{p_1}{T}-1)(e^\frac{E-p_1}{T}+1)} \\
        &\frac{1}{(1-e^{-\frac{p_3}{T}})(1+e^{-\frac{E-p_3}{T}})}\frac{p_1}{E^2 - p^2 - m_\psi^2} \frac{p_3^2(1-b)^2 + [p_4^2f+2p_3p_4(1-b)](f-\sqrt{f^2-c^2})}{p_4\sqrt{f^2-c^2}}
    \end{split}
\end{equation}
as a function of $\frac{m_\psi}{T}$.

```python
# @markdown ### Implement The Compton Integrand

# implement the integrand
@nb.cfunc(nb.types.double(nb.types.intc, nb.types.CPointer(nb.types.double)))
def integrand_compton(n, args):
    """Integrand for the four-dimensional integral for the (rescaled) interaction rate as
    a function of m/T"""
    mom3 = args[0]
    mom1 = args[1]
    tot_mom = args[2]
    tot_en = args[3]
    mass_over_temp_squared = args[4]
    boltzmann = args[5]

    mom4 = np.sqrt(
        tot_en ** 2 - mass_over_temp_squared - 2.0 * tot_en * mom3 + mom3 ** 2
    )

    cos_theta_1 = (
        -(tot_en ** 2) + mass_over_temp_squared + tot_mom ** 2 + 2.0 * tot_en * mom1
    ) / (2.0 * tot_mom * mom1)
    cos_theta_3 = (
        -(tot_en ** 2) + mass_over_temp_squared + tot_mom ** 2 + 2.0 * tot_en * mom3
    ) / (2.0 * tot_mom * mom3)
    cos_theta_4 = (
        tot_en ** 2 - mass_over_temp_squared + tot_mom ** 2 - 2.0 * tot_en * mom3
    ) / (2.0 * tot_mom * mom4)

    b_minus_1 = cos_theta_1 * cos_theta_3 - 1
    mom4_f = tot_en - mom3 - mom4 * cos_theta_1 * cos_theta_4
    mom4_sqrt_fsq_minus_csq = np.sqrt(
        mass_over_temp_squared * (1 - cos_theta_1 ** 2)
        + ((tot_en - mom3) * cos_theta_1 - mom4 * cos_theta_4) ** 2
    )
    phi_integral = (
        (mom3 * b_minus_1) ** 2
        + (mom4_f - 2 * mom3 * b_minus_1) * (mom4_f - mom4_sqrt_fsq_minus_csq)
    ) / mom4_sqrt_fsq_minus_csq

    constant = 1 / (8 * np.pi ** 4)
    if boltzmann:
        f_terms = np.exp(-tot_en)
    else:
        f_terms = 1 / (
            (np.exp(mom1) - 1)
            * (np.exp(tot_en - mom1) + 1)
            * (1 - np.exp(-mom3))
            * (1 + np.exp(mom3 - tot_en))
        )
    prefactor = mom1 / (tot_en ** 2 - tot_mom ** 2 - mass_over_temp_squared)

    return constant * f_terms * prefactor * phi_integral


# Define the integration limits of all four integrals
@nb.njit
def lim_tot_en_compton(mass_over_temp_squared, boltzmann):
    return [np.sqrt(mass_over_temp_squared), np.inf]


@nb.njit
def lim_tot_mom_compton(tot_en, mass_over_temp_squared, boltzmann):
    return [0, np.sqrt(tot_en ** 2 - mass_over_temp_squared)]


@nb.njit
def lim_mom1_compton(tot_mom, tot_en, mass_over_temp_squared, boltzmann):
    return [
        (tot_en ** 2 - mass_over_temp_squared - tot_mom ** 2)
        / (2.0 * (tot_en + tot_mom)),
        (tot_en ** 2 - mass_over_temp_squared - tot_mom ** 2)
        / (2.0 * (tot_en - tot_mom)),
    ]


@nb.njit
def lim_mom3_compton(mom1, tot_mom, tot_en, mass_over_temp_squared, boltzmann):
    return [
        (tot_en ** 2 - mass_over_temp_squared - tot_mom ** 2)
        / (2.0 * (tot_en + tot_mom)),
        (tot_en ** 2 - mass_over_temp_squared - tot_mom ** 2)
        / (2.0 * (tot_en - tot_mom)),
    ]
```

## Annihilation Process $\psi + \bar{\psi} \to a + \gamma$
We compute the annihilation process
\begin{equation}
    \begin{split}
        \Gamma_{\psi \bar{\psi} \to \gamma \phi} =& \frac{1}{n^{eq}_a}\frac{\alpha |\tilde{\epsilon}|^2}{32\pi^4}\int_{2m_\psi}^\infty dE\int_0^{\sqrt{E^2 - 4 m_\psi^2}} dp\int_{\frac{1}{2}\left|E\sqrt{1-\frac{4m_\psi^2}{s}} - p\right|}^{\frac{1}{2}\left(E\sqrt{1-\frac{4m_\psi^2}{s}} + p\right)} dp_1 \int_{\frac{E - p}{2}}^{\frac{E + p}{2}} dp_3 \\
        &\frac{1}{(e^\frac{E_1}{T}+1)(e^\frac{E-E_1}{T}+1)(1-e^{-\frac{p_3}{T}})(1-e^{-\frac{E-p_3}{T}})} \frac{(E^2 - p^2) p_1}{E_1} \\
        &\left(\frac{1}{p_3\sqrt{m_\psi^2\mathrm{sin}^2\theta_1 + (E_1\mathrm{cos}\theta_1 - p_1\mathrm{cos}\theta_3)^2}} + \frac{1}{p_4\sqrt{m_\psi^2\mathrm{sin}^2\theta_1 + (E_1\mathrm{cos}\theta_1 - p_1\mathrm{cos}\theta_4)^2}}\right)
    \end{split}
\end{equation}


```python
# @markdown ### Implement The Annihilation Integrand

# implement the integrand
@nb.cfunc(nb.types.double(nb.types.intc, nb.types.CPointer(nb.types.double)))
def integrand_annihilation(n, args):
    """Integrand for the four-dimensional integral for the (rescaled) interaction rate as
    a function of m/T"""
    mom3 = args[0]
    mom1 = args[1]
    tot_mom = args[2]
    tot_en = args[3]
    mass_over_temp_squared = args[4]
    boltzmann = args[5]

    en1 = np.sqrt(mass_over_temp_squared + mom1 ** 2)
    en2 = tot_en - en1
    mom4 = tot_en - mom3

    cos_theta_1 = (-(en2 ** 2) + mass_over_temp_squared + tot_mom ** 2 + mom1 ** 2) / (
        2.0 * tot_mom * mom1
    )
    cos_theta_3 = (-(tot_en ** 2) + tot_mom ** 2 + 2.0 * tot_en * mom3) / (
        2.0 * tot_mom * mom3
    )
    cos_theta_4 = (tot_en ** 2 + tot_mom ** 2 - 2.0 * tot_en * mom3) / (
        2.0 * tot_mom * mom4
    )

    denominator1_sq = (
        mass_over_temp_squared * (1 - cos_theta_1 ** 2)
        + (en1 * cos_theta_1 - mom1 * cos_theta_3) ** 2
    )
    denominator2_sq = (
        mass_over_temp_squared * (1 - cos_theta_1 ** 2)
        + (en1 * cos_theta_1 - mom1 * cos_theta_4) ** 2
    )

    phi_integral = 1 / (mom3 * np.sqrt(denominator1_sq)) + 1 / (
        mom4 * np.sqrt(denominator2_sq)
    )

    if boltzmann:
        f_terms = np.exp(-tot_en)
    else:
        f_terms = 1 / (
            (np.exp(en1) + 1)
            * (np.exp(en2) + 1)
            * (1 - np.exp(-mom3))
            * (1 - np.exp(-mom4))
        )
    mom_frac = (tot_en ** 2 - tot_mom ** 2) * mom1 / en1
    constant = 1.0 / (32 * np.pi ** 4)

    return constant * f_terms * mom_frac * phi_integral


# Define the integration limits of all four integrals
@nb.njit
def lim_tot_en_annihilation(mass_over_temp_squared, boltzmann):
    return [2.0 * np.sqrt(mass_over_temp_squared), np.inf]


@nb.njit
def lim_tot_mom_annihilation(tot_en, mass_over_temp_squared, boltzmann):
    return [0, np.sqrt(tot_en ** 2 - 4.0 * mass_over_temp_squared)]


@nb.njit
def lim_mom1_annihilation(tot_mom, tot_en, mass_over_temp_squared, boltzmann):
    part1 = (
        tot_en
        * np.sqrt(1 - 4.0 * mass_over_temp_squared / (tot_en ** 2 - tot_mom ** 2))
        / 2
    )
    part2 = tot_mom / 2
    return [np.abs(part1 - part2), part1 + part2]


@nb.njit
def lim_mom3_annihilation(mom1, tot_mom, tot_en, mass_over_temp_squared, boltzmann):
    return [(tot_en - tot_mom) / 2, (tot_en + tot_mom) / 2]
```

## Compute The Integral

```python
# @markdown ### Implement The Integral
# This program calculates the numerical integral of the axion production rate
# of the axion-fermion scattering process and fermion-fermion annihilation process
# under either quantum statistics or boltzmann statistics as a function of m/T

# job distributed to multiprocessers
def integrate(process, mass_over_temp, epsrel, epsabs, boltzmann):
    """Numerical integration of the integral to get the (rescaled) interaction rate as a function
    of (m/T)
    Parameters
    ----------
    process : string
        Can be one of 'compton', 'annihilation'
    mass_over_temp : float

    epsrel : float

    boltzmann : bool
        Set to True to allow boltzmann approximation
    Returns
    -------
    A list containing m/T, integral, abserr, neval, time, neval/s
    """
    if process == "compton":
        integrand = LowLevelCallable(integrand_compton.ctypes)
        lim = [
            lim_mom3_compton,
            lim_mom1_compton,
            lim_tot_mom_compton,
            lim_tot_en_compton,
        ]
    elif process == "annihilation":
        integrand = LowLevelCallable(integrand_annihilation.ctypes)
        lim = [
            lim_mom3_annihilation,
            lim_mom1_annihilation,
            lim_tot_mom_annihilation,
            lim_tot_en_annihilation,
        ]
    else:
        print("Error: wrong process name")
        return 0, 0, 0
    integral, abserr, out_dict = nquad(
        integrand,
        ranges=lim,
        args=(mass_over_temp ** 2, boltzmann),
        opts={"epsrel": epsrel, "epsabs": epsabs},
        full_output=True,
    )
    return integral, abserr, out_dict["neval"]


def work(process, mass_over_temp, epsrel, epsabs, boltzmann):
    t_start = time.time()
    # print(f'{time.ctime(t_start)}\tworking on m/T = {mass_over_temp:e}\n', end='', flush=True)
    integral, abserr, neval = integrate(
        process, mass_over_temp, epsrel, epsabs, boltzmann
    )
    t_end = time.time()
    result = [
        mass_over_temp,
        integral,
        abserr,
        neval,
        t_end - t_start,
        neval / (t_end - t_start),
    ]
    # print(time.ctime(t_end), '\tm/T = {:e}\tintegral = {:e}\tabserr = {:e}\tneval = {:e}\ttime = {:e}s\teval/sec = {:e}'.format(*result), flush=True)
    print(">", end="", flush=True)
    return result


def get_production_rate(
    process, mass_over_temp_list, epsrel=0.01, epsabs=-1, boltzmann=False, note=""
):
    """Calculate axion production rate as a function of m/T
    Parameters
    ----------
    process : string
        Can be one of 'compton', 'annihilation'
    mass_over_temp_list : list
        The list of m/T to work on.
    epsrel : float
        Relative tolerance for scipy.integrate.nquad (default = 1E-2)
    epsabs : float
        Absolute tolerance for scipy.integrate.nquad (default = -1)
    boltzmann : bool
        Set to True to allow boltzmann approximation
    note : string
        Attach a note to the result and output file

    Returns
    -------
    dict : Containing
        'process', 'epsrel', 'epsabs', 'boltzmann', 'note' - the input values
        ''
    """
    # print out the header
    print("-" * 80)
    print(f'{"# process": <16}= {process}')
    print(f'{"# boltzmann": <16}= {boltzmann}')
    print(f'{"# epsrel": <16}= {epsrel:e}')
    print(f'{"# epsabs": <16}= {epsabs:e}')
    print(f'{"# Note": <16} {note}')
    print(f'{"# Start at": <16}{time.ctime(time.time())}')
    print(f'{"# progress": <16}' + "=" * len(mass_over_temp_list))
    print(f'{"#": <16}', end="", flush=True)
    # generate arguments
    args = [
        [process, mass_over_temp, epsrel, epsabs, boltzmann]
        for mass_over_temp in mass_over_temp_list
    ]
    # distribute workload to threads
    with multiprocessing.Pool(NUM_THREAD) as p:
        result = p.starmap(work, args)
    # print out the result
    print()
    print(f'{"# Finish at": <16} {time.ctime(time.time())}')
    print(
        f'{"# m / T": <16}{"result": <16}{"abserr": <16}{"neval": <16}{"time (sec)": <16}eval / sec'
    )
    for row in result:
        for item in row:
            print(f"{item:e}\t", end="")
        print()
    result = np.array(result)
    out_dict = {
        "process": process,
        "epsrel": epsrel,
        "epsabs": epsabs,
        "boltzmann": boltzmann,
        "note": note,
        "mass_over_temp": result[:, 0],
        "integral": result[:, 1],
        "abserr": result[:, 2],
        "neval": result[:, 3],
        "time": result[:, 4],
        "neval_per_sec": result[:, 5],
    }
    return out_dict
```

## $g_*(T)$, $g_{*s}(T)$ homebrew

```python
# @markdown ### Get Estimated g\*, g\*s


def get_gstar_homebrew(particles=[]):
    class Particle:
        def __init__(self, name, mass, dof, isFermion):
            self.name = name
            self.mass = mass
            self.dof = dof
            self.isFermion = isFermion

    plist = []
    plist.append(Particle("y", 0, 2, False))
    plist.append(Particle("g", 0, 16, False))
    plist.append(Particle("W", 80.379, 6, False))
    plist.append(Particle("Z", 91.188, 3, False))
    plist.append(Particle("H", 125.1, 1, False))
    plist.append(Particle("t", 172.76, 12, True))
    plist.append(Particle("b", 4.18, 12, True))
    plist.append(Particle("c", 1.27, 12, True))
    plist.append(Particle("s", 93e-3, 12, True))
    plist.append(Particle("u", 2.16e-3, 12, True))
    plist.append(Particle("d", 4.67e-3, 12, True))
    plist.append(Particle("tau", 1776.86e-3, 4, True))
    plist.append(Particle("mu", 105.658e-3, 4, True))
    plist.append(Particle("e", 0.511e-3, 4, True))
    plist.append(Particle("nu", 0, 6, True))

    def integrand_gstar(p, m_over_T, pm_one):
        E = np.sqrt(m_over_T ** 2 + p ** 2)
        return E * p ** 2 / (np.exp(E) + pm_one)

    def gstar_single(T):
        gstar = 0
        for p in plist:
            factor = 15 / np.pi ** 4
            # QCD phase transition finishes
            if T < 0.12 and p.name in ["u", "d", "c", "s", "t", "b", "g"]:
                factor = 0
            # Remove SM particles with mass > Tcap
            if particles != [] and p.name not in particles:
                factor = 0

            m_over_T = p.mass / T
            if p.isFermion:
                pm_one = 1
            else:
                pm_one = -1
            gstar += (
                factor
                * p.dof
                * quad(integrand_gstar, 0, np.inf, args=(m_over_T, pm_one))[0]
            )
        return gstar

    Tmin = 1e-3
    Tmax = 10 ** 2.45
    T = np.exp(np.linspace(np.log(Tmin), np.log(Tmax), num=1000))
    gstar = [gstar_single(t) for t in T]
    gstar_interp_log = interp1d(np.log(T), gstar, kind="linear")

    def gstar_homebrew(T, deriv=0):
        if deriv == 0:
            return gstar_interp_log(np.log(T))
        elif deriv == 1:
            dT = 1e-3 * T
            return (
                gstar_interp_log(np.log(T + dT)) - gstar_interp_log(np.log(T - dT))
            ) / (2 * dT)
        else:
            print("error: deriv != 0 or 1")
            exit()

    return gstar_homebrew, gstar_homebrew


gstar_homebrew, gstar_s_homebrew = get_gstar_homebrew()
gstar_ynue, gstar_s_ynue = get_gstar_homebrew(particles=["y", "nu", "e"])
gstar_ynuemu, gstar_s_ynuemu = get_gstar_homebrew(particles=["y", "nu", "e", "mu"])
```

```python
# @markdown ### Plot g\*, g\*s
def plot_gstar():
    plt.figure(figsize=(5, 3))
    T = 10 ** np.linspace(-3, 2.45, num=100)
    plt.plot(T, [gstar_homebrew(t) for t in T], label=r"$g_*$ homebrew")
    plt.plot(T, [gstar_ynuemu(t) for t in T], label=r"$g_*$ ynuemu")
    plt.plot(T, [gstar_ynue(t) for t in T], label=r"$g_*$ ynue")
    plt.xscale("log")
    plt.xlabel("T (GeV)")
    plt.legend(loc="best")
    plt.show()


plot_gstar()
```

## $g_*(T)$, $g_{*s}(T)$ Borsanyi
Borsanyi et. al. arxiv:1606.07494

```python
# @markdown ### Get Borsanyi's g\*, g\*s


def get_gstar_borsanyi():
    interpolation_data = """
# data from arxiv 1606.07494 page 13
# interpolate these points with cubic spline gives 1% ~ 0.3% deviation
# Temperature in MeV
# log10(T)  gstar       gstar / gstar_s
0.000000	10.710000	1.002280
0.500000	10.740000	1.000290
1.000000	10.760000	1.000480
1.250000	11.090000	1.005050
1.600000	13.680000	1.021590
2.000000	17.610000	1.023240
2.150000	24.070000	1.054230
2.200000	29.840000	1.075780
2.400000	47.830000	1.061180
2.500000	53.040000	1.046900
3.000000	73.480000	1.017780
4.000000	83.100000	1.001230
4.300000	85.560000	1.003890
4.600000	91.970000	1.008870
5.000000	102.170000	1.007500
5.450000	104.980000	1.000230
#6.0         106.75      1.0
"""
    node_log10T = []
    node_gstar = []
    node_gstar_s = []
    lines = interpolation_data.split("\n")
    for line in lines:
        if line:
            words = line.split()
            if words[0][0] != "#":
                node_log10T.append(float(words[0]) - 3)  # Convrt from MeV to GeV
                node_gstar.append(float(words[1]))
                node_gstar_s.append(float(words[1]) / float(words[2]))

    f_gstar = CubicSpline(node_log10T, node_gstar)
    f_gstar_s = CubicSpline(node_log10T, node_gstar_s)

    node_log10T += [3.45, 4.45, 5.45]
    node_gstar += [106.75, 106.75, 106.75]
    node_gstar_s += [106.75, 106.75, 106.75]
    f_gstar_ext = CubicSpline(node_log10T, node_gstar)
    f_gstar_s_ext = CubicSpline(node_log10T, node_gstar_s)

    def check_temp_in_range(T, extended):
        T_min = 10 ** -3
        T_max = 10 ** 2.45
        if extended:
            T_max = 10 ** 5.45
        if isinstance(T, np.ndarray):
            assert all(T > T_min * 0.99) and all(T < T_max * 1.01)
        else:
            assert T > T_min * 0.99 and T < T_max * 1.01

    def gstar(T, deriv=0, extended=True):
        """
        T: float
            temperature in GeV, 1 MeV < T < 280 GeV
        deriv: int
            order of derivatrive. i.e. gstar'(T) = gstar(T, derive=1)
        returns g*(T)
        """
        check_temp_in_range(T, extended)
        if extended:
            return f_gstar_ext(np.log10(T), deriv)
        else:
            return f_gstar(np.log10(T), deriv)

    def gstar_s(T, deriv=0, extended=True):
        """
        T: float
            temperature in GeV, 1 MeV < T < 280 GeV
        deriv: int
            order of derivatrive. i.e. gstar'(T) = gstar(T, derive=1)
        returns g*_s(T)
        """
        check_temp_in_range(T, extended)
        if extended:
            return f_gstar_s_ext(np.log10(T), deriv)
        else:
            return f_gstar_s(np.log10(T), deriv)

    return gstar, gstar_s


gstar_borsanyi, gstar_s_borsanyi = get_gstar_borsanyi()
```

## $g_*(T)$, $g_{*s}(T)$ Saikawa
Saikawa and Shirai, arxiv 1803.01038

```python
# @markdown ### get Saikawa and Shirai's g\*, g\*s


def get_gstar_saikawa():
    # high temperature part: 120 Mev < T < 1E16 Gev
    coeff = np.array(
        [
            [1, 1.43382e-2, 1, 7.07388e1],
            [1.11724e0, 1.37559e-2, 6.07869e-1, 9.18011e1],
            [3.12672e-01, 2.92108e-03, -1.54485e-01, 3.31892e01],
            [-4.68049e-02, -5.38533e-04, -2.24034e-01, -1.39779e00],
            [-2.65004e-02, -1.62496e-04, -2.82147e-02, -1.52558e00],
            [-1.19760e-03, -2.87906e-05, 2.90620e-02, -1.97857e-02],
            [1.82812e-04, -3.84278e-06, 6.86778e-03, -1.60146e-01],
            [1.36436e-04, 2.78776e-06, -1.00005e-03, 8.22615e-05],
            [8.55051e-05, 7.40342e-07, -1.69104e-04, 2.02651e-02],
            [1.22840e-05, 1.17210e-07, 1.06301e-05, -1.82134e-05],
            [3.82259e-07, 3.72499e-09, 1.69528e-06, 7.83943e-05],
            [-6.87035e-09, -6.74107e-11, -9.33311e-08, 7.13518e-05],
        ]
    )
    a = coeff[:, 0]
    b = coeff[:, 1]
    c = coeff[:, 2]
    d = coeff[:, 3]

    def gstar_high_temp(T):
        t = np.log(T)
        numerator = 0
        denominator = 0
        for i in range(12):
            numerator += a[i] * t ** i
            denominator += b[i] * t ** i
        return numerator / denominator

    def gstar_s_high_temp(T):
        t = np.log(T)
        numerator = 0
        denominator = 0
        for i in range(12):
            numerator += c[i] * t ** i
            denominator += d[i] * t ** i
        ratio = 1 + numerator / denominator
        return gstar_high_temp(T) / ratio

    # low temperature part: T < 120 MeV
    m_e = 511e-6
    m_mu = 0.1056
    m_pi0 = 0.135
    m_pipm = 0.140
    m_1 = 0.5
    m_2 = 0.77
    m_3 = 1.2
    m_4 = 2.0

    def f_rho(x):
        return np.exp(-1.04855 * x) * (
            1 + 1.03757 * x + 0.508630 * x ** 2 + 0.0893988 * x ** 3
        )

    def b_rho(x):
        return np.exp(-1.03149 * x) * (
            1 + 1.03317 * x + 0.398264 * x ** 2 + 0.0648056 * x ** 3
        )

    def f_s(x):
        return np.exp(-1.04190 * x) * (
            1 + 1.03400 * x + 0.456426 * x ** 2 + 0.0595248 * x ** 3
        )

    def b_s(x):
        return np.exp(-1.03365 * x) * (
            1 + 1.03397 * x + 0.342548 * x ** 2 + 0.0506182 * x ** 3
        )

    def s_fit(x):
        return 1 + 7 / 4 * np.exp(-1.0419 * x) * (
            1 + 1.034 * x + 0.456426 * x ** 2 + 0.0595249 * x ** 3
        )

    def gstar_low_temp(T):
        return (
            2.030
            + 1.353 * s_fit(m_e / T) ** (4 / 3)
            + 3.495 * f_rho(m_e / T)
            + 3.446 * f_rho(m_mu / T)
            + 1.05 * b_rho(m_pi0 / T)
            + 2.08 * b_rho(m_pipm / T)
            + 4.165 * b_rho(m_1 / T)
            + 30.55 * b_rho(m_2 / T)
            + 89.4 * b_rho(m_3 / T)
            + 8209 * b_rho(m_4 / T)
        )

    def gstar_s_low_temp(T):
        return (
            2.008
            + 1.923 * s_fit(m_e / T)
            + 3.442 * f_s(m_e / T)
            + 3.468 * f_s(m_mu / T)
            + 1.034 * b_s(m_pi0 / T)
            + 2.068 * b_s(m_pipm / T)
            + 4.16 * b_s(m_1 / T)
            + 30.55 * b_s(m_2 / T)
            + 90 * b_s(m_3 / T)
            + 6209 * b_s(m_4 / T)
        )

    def derivative(func):
        # return d func(T) / d log10(T)
        def result(T):
            t = np.log(T)
            dt = 1e-6
            t1 = np.exp(t - dt)
            t2 = np.exp(t + dt)
            y1 = func(t1)
            y2 = func(t2)
            dydlogT = (y2 - y1) / (2 * dt)
            return dydlogT * np.log(10)

        return result

    # combine high temp and low temp
    def gstar_single(T, deriv=0):
        if T < 0.12:
            if deriv:
                return derivative(gstar_low_temp)(T)
            else:
                return gstar_low_temp(T)
        else:
            if deriv:
                return derivative(gstar_high_temp)(T)
            else:
                return gstar_high_temp(T)

    def gstar_s_single(T, deriv=0):
        if T < 0.12:
            if deriv:
                return derivative(gstar_s_low_temp)(T)
            else:
                return gstar_s_low_temp(T)
        else:
            if deriv:
                return derivative(gstar_s_high_temp)(T)
            else:
                return gstar_s_high_temp(T)

    def gstar(T, deriv=0):
        if isinstance(T, np.ndarray):
            return np.array([gstar_single(T[i], deriv=deriv) for i in range(len(T))])
        else:
            return gstar_single(T, deriv=deriv)

    def gstar_s(T, deriv=0):
        if isinstance(T, np.ndarray):
            return np.array([gstar_s_single(T[i], deriv=deriv) for i in range(len(T))])
        else:
            return gstar_s_single(T, deriv=deriv)

    return gstar, gstar_s


gstar_saikawa, gstar_s_saikawa = get_gstar_saikawa()
```

## Running Couplings (Ben's Notebook)
Providing the running Couplings for the Standard Model

### Running Coupling Functions

We need the coupling strengths as a function of temperature. Here, we use the values as provided
by the 1-loop renormalization group evolution. In general, we have
$\alpha(\mu)^{-1} = \alpha(M)^{-1} - 2 b_0 \ln\left(\frac{\mu}{M}\right)$,
or
$g(T)^{-2} = \frac{1}{4\pi\alpha(M_Z)} - \frac{b_0}{2\pi} \ln\left(\frac{T}{M_Z}\right)$,
where $b_0$ is the coefficient of the 1-loop beta-function.

For the strong/$SU(3)$ coupling $g_s = g_3$, we have
$b_0^s = -\frac{11N_c - 2N_f}{12\pi} = -\frac{21}{12\pi} = -\frac{7}{4\pi}$.
Hence, the running strong coupling is:
$g_3(T) = \left[ \frac{1}{4\pi\alpha_s(M_Z)} + \frac{11N_c - 2N_f}{24\pi^2}
\ln\left(\frac{T}{M_Z}\right) \right]^{\!-1/2}$:

```python
# @markdown #### Implement g3
def g_3(temp):
    """Strong coupling parameter as described by the 1-loop renormalization group evolution as
    a function of temperature

    Parameters
    ----------
    temp : float
        Temperature (in GeV)

    Returns
    -------
    float
        Strong coupling constant
    """
    m_z = 91.1876  # mass of Z-boson in GeV
    alpha3_at_z = 0.1185  # alpha_s(M_Z)
    n_c = 3  # number of colors
    n_f = 6  # number of quarks
    return 1.0 / np.sqrt(
        1.0 / (4 * np.pi * alpha3_at_z)
        + (11 * n_c - 2 * n_f) / (24 * np.pi ** 2) * np.log(temp / m_z)
    )
```

In the case of the $SU(2)$ gauge coupling $g_2$, the 1-loop $\beta$-function is
$b_0 = -\frac{11N - 2 N_f - N_s/2}{12\pi} = -\frac{19}{24\pi}$.
Therefore, we have
$g_2(T) = \left[ \frac{1}{4\pi\alpha_2(M_Z)} + \frac{11N - 2N_f - N_s/2}{24\pi^2}
\ln\left(\frac{T}{M_Z}\right) \right]^{\!-1/2}$:

```python
# @markdown #### Implement g2
def g_2(temp):
    """SU(2) coupling parameter as described by the 1-loop renormalization group evolution as
    a function of temperature

    Parameters
    ----------
    temp : float
        Temperature (in GeV)

    Returns
    -------
    float
        Strong coupling constant
    """
    m_z = 91.1876  # mass of Z-boson in GeV
    alpha_em_at_z = 1 / 127.934  # fine-structure constant alpha_EM(M_Z)
    weinberg_angle_at_z = 0.23126  # Weinberg angle sin^2(theta_w)
    alpha2_at_z = alpha_em_at_z / weinberg_angle_at_z  # alpha_w(M_Z)
    n_su = 2
    n_f = 6
    n_s = 1
    return 1.0 / np.sqrt(
        1.0 / (4 * np.pi * alpha2_at_z)
        + (11 * n_su - 2 * n_f - n_s / 2.0) / (24 * np.pi ** 2) * np.log(temp / m_z)
    )
```

The $\beta$-function for the $U(1)$ (hypercharge) coupling $g_1$ ($g_Y$) is
$b_0 = \frac{41}{40\pi}$ with SU(5) normalization ($b_0 = \frac{41}{24\pi}$). The running coupling
is therefore given by
$g_1(T) = \left[ \frac{1}{4\pi\alpha_1(M_Z)} - \frac{41}{80\pi^2}
\ln\left(\frac{T}{M_Z}\right) \right]^{\!-1/2}$:

```python
# @markdown #### Implement g1
def g_1(temp):
    """U(1) coupling parameter as described by the 1-loop renormalization group evolution
    (with SU(5) normalization) as a function of temperature

    Parameters
    ----------
    temp : float
        Temperature (in GeV)

    Returns
    -------
    float
        Strong coupling constant
    """
    m_z = 91.1876  # mass of Z-boson in GeV
    alpha_em_at_z = 1 / 127.934  # fine-structure constant alpha_EM(M_Z)
    weinberg_angle_at_z = 0.23126  # Weinberg angle sin^2(theta_w)
    alpha1_at_z = (
        5.0 / 3.0 * alpha_em_at_z / (1 - weinberg_angle_at_z)
    )  # alpha_1(M_Z) (in SU(5) norm.)
    return 1.0 / np.sqrt(
        1.0 / (4 * np.pi * alpha1_at_z) - 41.0 / (80 * np.pi ** 2) * np.log(temp / m_z)
    )
```

```python
# @markdown #### Implement gY
def g_y(temp):
    """Hypercharge coupling parameter as described by the 1-loop renormalization group evolution as
    a function of temperature

    Parameters
    ----------
    temp : float
        Temperature (in GeV)

    Returns
    -------
    float
        Strong coupling constant
    """
    m_z = 91.1876  # mass of Z-boson in GeV
    alpha_em_at_z = 1 / 127.934  # fine-structure constant alpha_EM(M_Z)
    weinberg_angle_at_z = 0.23126  # Weinberg angle sin^2(theta_w)
    alpha_y_at_z = alpha_em_at_z / (
        1 - weinberg_angle_at_z
    )  # alpha_Y(M_Z) (in SU(5) normalization)
    return 1.0 / np.sqrt(
        1.0 / (4 * np.pi * alpha_y_at_z) - 41.0 / (48 * np.pi ** 2) * np.log(temp / m_z)
    )
```

### Running $\mathbf{\alpha}_\mathbf{i}$

```python
# @markdown Implement alpha
def alpha(temp, group="U(1)"):
    """Coupling constant alpha_i as described by the 1-loop renormalization group evolution as a
    function of temperature for hypercharge, U(1), SU(2) and SU(3) of the Standard Model

    Parameters
    ----------
    temp : float
        Temperature (in GeV)

    group : string
        Standard Model gauge group: 'U(1)', 'hypercharge', 'em', 'SU(2)' or 'SU(3)'

    Returns
    -------
    float
        Coupling constant
    """
    if group == "U(1)":
        result = g_1(temp) ** 2 / (4.0 * np.pi)
    elif group == "hypercharge":
        result = g_y(temp) ** 2 / (4.0 * np.pi)
    elif group == "em":
        warnings.warn(
            "Something not quite right: alpha(m_Z) correct, but not alpha(T to 0)..."
        )
        result = (
            g_y(temp) ** 2
            * g_2(temp) ** 2
            / (g_y(temp) ** 2 + g_2(temp) ** 2)
            / (4.0 * np.pi)
        )
    elif group == "SU(2)":
        result = g_2(temp) ** 2 / (4.0 * np.pi)
    elif group == "SU(3)":
        result = g_3(temp) ** 2 / (4.0 * np.pi)
    else:
        raise ValueError(
            "'group' has to be 'U(1)', 'hypercharge', 'em', 'SU(2)' or 'SU(3)'."
        )

    return result
```

Examine $\alpha_\textrm{em}$ at mass of Z-boson, muon, electron

## $\Delta N_\textrm{eff}$ vs $\Lambda_\psi$

### Preliminaries

```python
# @markdown ### Implement Neff
nbin = 100
use_custom_integration = False  # @param {type:'boolean'}
sim_min_step_num = 1e2
sim_max_step_num = 1e4
sim_acc = 0.1


def get_f_integral_loglog(pr):
    f_integral_loglog = interp1d(
        np.log(pr["mass_over_temp"]), np.log(pr["integral"]), kind="cubic"
    )
    return f_integral_loglog


def get_f_integral_loglog_linear(pr):
    xs = np.log(pr["mass_over_temp"])
    ys = np.log(pr["integral"])

    def f_integral_loglog(x):
        return np.interp(x, xs, ys)

    return f_integral_loglog


def f_Yaeq(x, mass):
    T = mass / x
    return 45 * ZETA_3 / (2 * np.pi ** 4 * gstar_s_tmp(T))


def f_B(x, lambda_, mass, A_psi):
    T = mass / x
    prefactor = (
        4 * np.pi * np.sqrt(90) * M_PL * A_psi(T) * mass / (ZETA_3 * lambda_ ** 2)
    )
    part_g = (
        1 + gstar_s_tmp(T, deriv=1) / (3 * gstar_s_tmp(T) * np.log(10))
    ) / gstar_tmp(T) ** 0.5
    return prefactor * part_g * np.exp(f_integral_loglog(np.log(x)))


def f_A(x, lambda_, mass, A_psi):
    return f_B(x, lambda_, mass, A_psi) * f_Yaeq(x, mass)


def f_C(y, x, lambda_, mass, A_psi):
    result = quad(
        f_B, y, x, args=(lambda_, mass, A_psi), epsabs=1e-4, epsrel=-1, limit=500
    )
    return result[0]


def f_C_custom(y, x, lambda_, mass, A_psi):
    result = 0
    step_size = (np.log(x) - np.log(y)) / nbin
    for i in range(nbin):
        t = np.exp(np.log(y) + (i + 0.5) * step_size)
        result += step_size * t * f_B(t, lambda_, mass, A_psi)
    return result


def integrand_Ya(y, x, lambda_, mass, A_psi):
    return f_A(y, lambda_, mass, A_psi) * np.exp(-f_C(y, x, lambda_, mass, A_psi))


def integrand_Ya_custom(y, x, lambda_, mass, A_psi):
    return f_A(y, lambda_, mass, A_psi) * np.exp(
        -f_C_custom(y, x, lambda_, mass, A_psi)
    )


def f_Ya(x, lambda_, mass, A_psi, T_i):
    result = quad(
        integrand_Ya,
        mass / T_i,
        x,
        args=(x, lambda_, mass, A_psi),
        epsabs=1e-4,
        epsrel=-1,
        limit=500,
    )
    return result[0]


def f_Ya_custom(x, lambda_, mass, A_psi, T_i):
    result = 0
    x0 = mass / T_i
    step_size = (np.log(x) - np.log(x0)) / nbin
    for i in range(nbin):
        t = np.exp(np.log(x0) + (i + 0.5) * step_size)
        result += step_size * t * integrand_Ya_custom(t, x, lambda_, mass, A_psi)
    return result


# solving for Ya in the simulation way (accumulating Ya_prime) instead of using the general solution
def f_dYa_sim(x, lambda_, mass, A_psi, Ya):
    return x * (f_A(x, lambda_, mass, A_psi) - f_B(x, lambda_, mass, A_psi) * Ya)


def f_Ya_sim(x, lambda_, mass, A_psi, T_i):
    x0 = mass / T_i
    max_step_size = (np.log(x) - np.log(x0)) / sim_min_step_num
    min_step_size = (np.log(x) - np.log(x0)) / sim_max_step_num
    step_size = max_step_size

    Ya = 0
    t = x0
    max_abs_error = 0
    finished = False
    while not finished:
        # deal with the last step
        if t * np.exp(step_size) > x:
            step_size = np.log(x) - np.log(t)
            finished = True
        # calculate next step with runge-kuta addaptive step size
        y1 = Ya + f_dYa_sim(t, lambda_, mass, A_psi, Ya) * step_size
        y2 = Ya + f_dYa_sim(t, lambda_, mass, A_psi, Ya) * step_size * 0.5
        t2 = t * np.exp(step_size * 0.5)
        y2 = y2 + f_dYa_sim(t2, lambda_, mass, A_psi, y2) * step_size * 0.5
        second_order_error = 2 * (y2 - y1)
        # register the step if the accuracy goal is met
        # or if the absolute error is small compared to some earlier large error
        if abs(second_order_error / (y1 - Ya)) < sim_acc or abs(
            second_order_error
        ) / step_size < max_abs_error / (max_step_size * sim_min_step_num):
            Ya = y1 + second_order_error
            t *= np.exp(step_size)
            max_abs_error = max(max_abs_error, abs(second_order_error))
            if step_size * 2 < max_step_size:
                step_size *= 2
        # reduce the step size when the goal is not met
        elif step_size * 0.5 > min_step_size:
            step_size *= 0.5
        # register a step when the step size has already reached the minimum
        else:
            Ya = y1 + second_order_error
            t *= np.exp(step_size)
        # when the step size is overlarge but has already reached the minimum
        # the process might not converge, in that case we manually set
        # Ya = Yaeq
        if Ya > f_Yaeq(t, mass) or Ya < 0 or np.isnan(Ya):
            Ya = f_Yaeq(t, mass)
    return Ya


def f_delta_Neff(lambda_, mass, A_psi, T_i, T_f):
    Ya = f_Ya(mass / T_f, lambda_, mass, A_psi, T_i)
    # check whether the integral was bad
    # the simulation method is fast and can be a reference
    global sim_max_step_num, sim_acc
    sim_max_step_num = 1e4
    sim_acc = 0.1
    Ya_sim = f_Ya_sim(mass / T_f, lambda_, mass, A_psi, T_i)
    if abs(Ya / Ya_sim - 1) > 0.1:
        # Ya = f_Yaeq(mass/T_f, mass)
        sim_max_step_num = 1e6
        sim_acc = 0.01
        Ya = f_Ya_sim(mass / T_f, lambda_, mass, A_psi, T_i)
    return 74.8497 * Ya ** (4 / 3)


def f_delta_Neff_custom(lambda_, mass, A_psi, T_i, T_f):
    return 74.8497 * f_Ya_custom(mass / T_f, lambda_, mass, A_psi, T_i) ** (4 / 3)


def f_delta_Neff_sim(lambda_, mass, A_psi, T_i, T_f):
    return 74.8497 * f_Ya_sim(mass / T_f, lambda_, mass, A_psi, T_i) ** (4 / 3)


def work(lambda_, mass, A_psi, T_i, T_f):
    if use_custom_integration:
        result = f_delta_Neff_custom(lambda_, mass, A_psi, T_i, T_f)
    elif use_simulation_method:
        result = f_delta_Neff_sim(lambda_, mass, A_psi, T_i, T_f)
    else:
        result = f_delta_Neff(lambda_, mass, A_psi, T_i, T_f)
    print(">", end="", flush=True)
    return result


def get_dNeff(lambda_list, mass, A_psi, T_i, T_f):
    n_point = len(lambda_list)
    args = [[lambda_list[i], mass, A_psi, T_i, T_f] for i in range(n_point)]

    # work and show progress
    print("=" * n_point)
    with multiprocessing.Pool(NUM_THREAD) as p:
        result = p.starmap(work, args)
    print()

    # show the report
    print("# delta_Neff vs the axion-fermion-fermion coupling energy lambda (in GeV)")
    print(f"# T_i, T_f = {T_i:e},\t{T_f:e}")
    print(f'{"# lambda": <16}delta_Neff')
    for i in range(n_point):
        print(f"{lambda_list[i]:e}\t{result[i]:e}")

    return result


def A_psi_em(T):
    return ALPHA
    # return alpha(T, group='em')


def A_psi_s(T):
    # return 4 * alpha(T, group='SU(3)')
    if T > 0.66818:
        return 4 * alpha_s(T)
    else:
        return 4
```
